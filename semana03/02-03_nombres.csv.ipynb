```python
import pandas as pd
import numpy as np
```

```python
# Leer el archivo nombres.csv y convertirlo en un df
df_base = pd.read_csv('nombres.csv')
# Ver los primeros registros del df
df_base.head()
```

```python
# Mostrar información sobre el df creado
df_base.info()
```

```python
#Filtrar todas los hombres con nombre Mary
#Crear un nuevo dataframe
df_mary = df_base[(df_base["sex"] == "M") & (df_base["name"] == "Mary")]
df_mary
```

```python
#Ya no necesitamos el sexo y el nombre, hacemos una copia
nuevo_df = df_mary[["state", "quantity"]]
nuevo_df.head()
```

```python
#Obtener la suma de la columna quantity
nuevo_df[["quantity"]].sum()
```

```python
estado_df = nuevo_df.groupby("state").agg("sum")
estado_df
```

```python
# Reto:
# Obtener la cantidad promedio de Mary por estado usando estados_df
promedio_df = nuevo_df.groupby("state").agg("mean")
promedio_df
```

```python
# Ordenar el set de datos por la mayor cantidad
estado_df.sort_values('quantity', ascending=False)
```

```python
# Reto
# Obtener los 10 registros con la mayor cantidad de Mary
estado_df.sort_values('quantity', ascending=False).head(10)
```

```python
# Para convertir el índice en una columna del df, se puede generar uno nuevo aleatorio con reset_index
# Nota al uso de inplace=True para modificar el df
estado_df.reset_index(inplace=True)
estado_df
```

```python
estado_df.to_excel('estados_mary.xlsx', index=None)
```
